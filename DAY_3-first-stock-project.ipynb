```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import pandas_market_calendars as mcal
from datetime import datetime, timedelta
import holidays

nyse = mcal.get_calendar("NYSE")
us_holidays = holidays.UnitedStates()

def next_trading_day(date_str):
    date = datetime.strptime(date_str, "%Y-%m-%d")
    while date.weekday() >= 5 or date in us_holidays:
        date += timedelta(days=1)
    return date.strftime("%Y-%m-%d")

def dateconvert(date):
    year, month, day = map(int, date.split("-"))
    return year, month, day

while True:
    ticker_input = input(
        "Enter the stock symbol(s) / ticker symbol(s) of your preferred stock (e.g. MSFT).\n"
        "For multiple stocks, separate each ticker with a comma (e.g., MSFT, AAPL, GOOG (Up to 3)): "
    )
    if not ticker_input.replace(".", "").replace(",", "").replace(" ", "").isalnum():
        print("Ticker should only contain letters.")
        continue
    tickers = [t.strip().upper() for t in ticker_input.split(",")]
    if len(tickers) > 3:
        print("You can only compare three stocks in one instance.")
        continue
    ticker_fullnames = {}
    all_valid = True
    for t in tickers:
        try:
            name = yf.Ticker(t).info.get("longName")
        except Exception:
            print(f"Ticker '{t}' not found. Try again.")
            all_valid = False
        else:
            if not name:
                print(f"Ticker '{t}' not found. Try again.")
                all_valid = False
            else:
                ticker_fullnames[t] = name
    if all_valid:
        break

while True:
    start_date = input("Enter the starting date (YYYY-MM-DD format): ")
    end_date = input("Enter the ending date (YYYY-MM-DD format): ")
    try:
        syear, smonth, sdate = dateconvert(start_date)
        eyear, emonth, edate = dateconvert(end_date)
        start_dt = datetime.strptime(start_date, "%Y-%m-%d")
        end_dt = datetime.strptime(end_date, "%Y-%m-%d")
    except ValueError:
        print("Invalid date format; use YYYY-MM-DD.")
        continue
    if start_dt.weekday() >= 5 or start_dt in us_holidays:
        print(f"{start_date} is a weekend or holiday. Please re-enter.")
        continue
    if end_dt.weekday() >= 5 or end_dt in us_holidays:
        print(f"{end_date} is a weekend or holiday. Please re-enter.")
        continue
    if start_dt == end_dt:
        print("Starting date and ending date cannot be the same.")
        continue
    if start_dt > end_dt:
        print("Starting date cannot be after the ending date.")
        continue
    if end_dt > datetime.now():
        print("Ending date must not exceed the current date.")
        continue
    break

print()

days_gap = (end_dt - start_dt).days
if days_gap <= 10:
    rownum = 5
elif days_gap <= 30:
    rownum = 10
else:
    rownum = 15

if start_dt >= end_dt:
    end_dt = start_dt + timedelta(days=1)
    end_date = end_dt.strftime("%Y-%m-%d")

if days_gap <= 5:
    interval = "15m"
    interval_name = "15 minute"
elif days_gap <= 30:
    interval = "1h"
    interval_name = "1 hour"
elif days_gap <= 90:
    interval = "1d"
    interval_name = "1 day"
elif days_gap <= 365:
    interval = "1wk"
    interval_name = "1 week"
else:
    interval = "1mo"
    interval_name = "1 month"

all_data = {}
for t in tickers:
    try:
        df = yf.download(t, start=start_date, end=end_date, interval=interval, auto_adjust=True)
        if df.empty:
            print(f"No data found for {t}. Skipping.")
            continue
        all_data[t] = df
    except Exception as e:
        print(f"Error fetching {t}: {e}")

if not all_data:
    print("No data found for the given date range.")
else:
    metrics = ['Close', 'High', 'Low', 'Open', 'Volume']
    for metric in metrics:
        print(f"\n{metric.upper()}\n")
        df_metric = pd.concat([df[[metric]].rename(columns={metric: t}) for t, df in all_data.items()], axis=1)
        print(df_metric)
        print()
        df_metric.plot(title=f"{metric} ({interval_name} interval)")
        plt.show()
```
